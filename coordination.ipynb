# Calculate the coordination metrics of Wikipedians. 
To calculate the coordination, I look into talk pages in Wikipedia. How it's done is explained throughout the notebook.

````python
DATA_PATH = "/dlabdata1/turkish_wiki"
````

````python
revision_df = pd.read_csv(f'{DATA_PATH}/revision_events.tsv.gz', sep="\t", usecols= ['event_type', 'page_namespace',  'event_entity', 'event_type', 'event_timestamp', 
       'event_user_id', 'event_user_groups', 'event_user_revision_count',
       'event_user_seconds_since_previous_revision', 'page_id', 
       'event_user_id', 'event_user_is_bot_by',
       'page_title', 'page_revision_count', 'revision_minor_edit',
       'revision_text_bytes', 'revision_text_bytes_diff','revision_is_identity_revert',
       'revision_is_identity_revert', 'revision_is_identity_reverted'], error_bad_lines=False, warn_bad_lines=True, compression = 'gzip')
````

````python
revision_df
````

````python
# Mask to obtain revisions coming from registered accounts.
account_mask = (~revision_df.event_user_id.isna()) & (revision_df.event_user_is_bot_by.isna())
# Mask to obtain revisions coming from bots.
bot_mask = (~revision_df.event_user_is_bot_by.isna())
# Mask to obtain revisions coming from anonymous users.
anonymous_mask = (revision_df.event_user_id.isna()) & (revision_df.event_user_is_bot_by.isna())
````

````python
# Create a new column user_kind and put the corresponding user kinds for each row of the df.
revision_df['user_kind'] = np.nan
revision_df.loc[anonymous_mask, 'user_kind'] = 'anonymous'
revision_df.loc[bot_mask, 'user_kind'] = 'bot'
revision_df.loc[account_mask, 'user_kind'] = 'account'
````

````python
revision_df
````

````python
def get_raw_revision_data(revision_df, namespace, file_name):
    '''
        Creates a DataFrame containing all revisions of the selected namespace.
    '''
    try:
        df = revision_df.convert_dtypes()
        create_revision_mask = (df.event_entity=='revision') & (df.event_type == 'create')
        
        if type(namespace) == list:
            ns_mask = df.page_namespace.isin(namespace)
        else:
            ns_mask = df.page_namespace == namespace
            
        df = df[create_revision_mask & ns_mask]
        df['event_timestamp'] = pd.to_datetime(df['event_timestamp'],  utc = True, errors = 'coerce')
        df.to_csv(f'{DATA_PATH}/processed_data/{file_name}.csv')
        return df

    except Exception as e:
        logging.error(f'Error: {str(e)}')
        return None
````

````python
def get_processed_revision_data(raw_df, file_name):
    
    # Get Daily revisions by user_type
    raw_df.groupby([pd.Grouper(key = 'event_timestamp', freq='D'), 'user_kind'])\
          .agg({'event_user_id': 'size', 'revision_text_bytes': 'sum'}).reset_index()\
          .rename(columns = {'event_timestamp': 'date', 'event_user_id': 'number_of_edits', 
                             'revision_text_bytes': 'total_edited_bytes'})\
            .to_csv(f'{DATA_PATH}/processed_data/daily_{file_name}.csv')
    
    print(f"DataFrame saved at {DATA_PATH}/processed_data/daily_{file_name}.csv")
    
    # Get Daily revisions by user_type and page_id
    raw_df.groupby([pd.Grouper(key = 'event_timestamp', freq='D'), 'page_id', 'user_kind'])\
          .agg({'event_user_id': 'size', 'revision_text_bytes': 'sum'}).reset_index()\
          .rename(columns = {'event_timestamp': 'date', 'event_user_id': 'number_of_edits', 
                             'revision_text_bytes': 'total_edited_bytes'})\
            .to_csv(f'{DATA_PATH}/processed_data/daily_{file_name}_by_page_id.csv')
    
    print(f"DataFrame saved at {DATA_PATH}/processed_data/daily_{file_name}_by_page_id.csv")

    # Get raw revisions by registered accounts
    raw_df[raw_df['user_kind'] == 'account'][['event_timestamp', 'event_user_id','page_id', 'revision_minor_edit',
       'revision_text_bytes', 'revision_text_bytes_diff',
       'revision_is_identity_reverted', 'revision_is_identity_revert']].to_csv(f'{DATA_PATH}/processed_data/{file_name}_by_registered_users.csv')
    
    print(f"DataFrame saved at {DATA_PATH}/processed_data/{file_name}_by_registered_users.csv")

````

## 1) Article Talk pages
Wikipedia editors collaborate with eachother through editions of Talk pages. Editors can discuss improvements to articles or other Wikipedia pages. The talk page associated with an article is named "Talk:Example", where "Example" is the name of the article. We will only conduct the analysis for the Talk pages of Wikipedia articles and not go into the detail to analyse the Talk pages to pages like categories or users.

The namespace corresponding to talk pages is ```1```.

````python
talk = get_raw_revision_data(revision_df, namespace = 1, file_name = 'talk_revisions')
````

The above cell saves the Talk page revisions in raw format.

````python
get_processed_revision_data(talk, 'talk_revisions')
````
